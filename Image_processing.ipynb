```python
import plotly.express as px
import cv2
import numpy as np
import matplotlib.pyplot as plt
import json
import pandas as pd
import base64
from PIL import Image
from collections import Counter
from skimage import io
from skimage import color
from skimage.transform import resize
from scipy.signal import convolve2d
from google.colab.patches import cv2_imshow
```

```python
image=cv2.imread("/content/Snoop Dogg.png")
print("Original Image")
cv2_imshow(image)
img = cv2.imread('/content/Snoop Dogg.png')
gray=np.dot(img[...,:3],[0.3,0.59,0.11])
cv2_imshow(gray)
cv2.imwrite("GRAY_IMAGE.jpg", gray)
plt.axis("off")
```

```python
grayscale_array = np.array(gray)
flattened_grayscale = grayscale_array.flatten()
data_for_task_2 = [{"pixel_value": int(value)} for value in flattened_grayscale]
```

```python
import csv
default_filename = "pixel_values.csv"
filename = default_filename
with open(filename, 'w', newline='') as csvfile:
    fieldnames = ["pixel_value"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for entry in data_for_task_2:
        writer.writerow(entry)
print(f'Data saved to {filename}')
```

```python
gray_image_path = "/content/Snoop Dogg.png"
gray = Image.open(gray_image_path)
grayscale_array = np.array(gray)
flattened_array = grayscale_array.flatten()
histogram, bins = np.histogram(flattened_array, bins=256, range=[0, 256])
cumulative_sum = np.cumsum(histogram)
min_val = cumulative_sum.min()
max_val = cumulative_sum.max()
normalized_histogram = ((cumulative_sum - min_val) * 255) / (max_val - min_val)
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.plot(histogram)
plt.title('Histogram')
plt.subplot(122)
plt.plot(normalized_histogram)
plt.title('Normalized Histogram')
plt.show()
```
